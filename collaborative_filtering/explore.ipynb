```julia
using DataFrames
using CSV
using Random
using Distributions
using Plots
```

```julia
# plot a normal distribution centered around 5
n = Normal(5, 3)
x = LinRange(0, 10, 1000)
y = pdf.(n, x)
plot(x, y, label="normal distribution")
```

```julia
# plot a Inverse Gaussian distribution centered around 5 - this will be likened to the distribution of prices
n = InverseGaussian(300, 200)
x = LinRange(0, 1000, 1000)
y = pdf.(n, x)
plot(x, y, label="Inverse Gaussian")
```
